# EMI Pre-Compliance — Meeting Deck

Single scrollable view of every analysed mode. All plots are loaded from
`reports/*.html` (regenerated by `emi_analysis.ipynb` /
`run_all_modes.py`), so this notebook never re-executes the heavy analysis —
just run the setup cell once and scroll.

**Sections per mode**
1. Overlay of all antenna positions vs CISPR 11
2. Worst-case envelope vs CISPR 11 limit
3. Δ vs. the mode's baseline trace (if one is defined)

The final section overlays the worst-case envelopes of every mode against
each other.

```python
from pathlib import Path
from IPython.display import IFrame, Markdown, display

REPORTS = Path('reports')
IFRAME_W, IFRAME_H = '100%', 620

# Keep this list in the order you want to present.
# Each entry: (display title, slug used in reports/*.html, baseline label or None)
MODES = [
    ('Standby (18 May)',                'standby',                              None),
    ('Movement + X-ray (19 May)',       'movement_xray',                        None),
    ('Configurations sweep (20 May)',   'configurations_20_may',                'Only cabinets'),
    ('CLEA SAMD sweep (20 May)',        'clea_samd_sweep_20_may',               'Only cabinets'),
    ('Ambient vs Only cabinets',        'ambient_vs_only_cabinets',             'Ambient (no system)'),
]

def show_mode(title, slug, baseline):
    display(Markdown(f'## {title}'))
    overlay  = REPORTS / f'overlay_positions_{slug}.html'
    envelope = REPORTS / f'envelope_vs_cispr11_{slug}.html'
    delta    = REPORTS / f'baseline_delta_{slug}.html'

    display(Markdown(f'**1. Positions overlay** — [{overlay.name}]({overlay.as_posix()})'))
    display(IFrame(overlay.as_posix(), width=IFRAME_W, height=IFRAME_H))

    display(Markdown(f'**2. Envelope vs CISPR 11** — [{envelope.name}]({envelope.as_posix()})'))
    display(IFrame(envelope.as_posix(), width=IFRAME_W, height=IFRAME_H))

    if baseline and delta.exists():
        display(Markdown(
            f'**3. Δ vs. baseline `{baseline}`** — [{delta.name}]({delta.as_posix()})'
        ))
        display(IFrame(delta.as_posix(), width=IFRAME_W, height=IFRAME_H + 80))
    elif baseline:
        display(Markdown(f'_Δ plot missing: `{delta}` not found._'))
```

# Per-mode results

```python
show_mode(*MODES[0])  # Standby
```

```python
show_mode(*MODES[1])  # Movement + X-ray
```

```python
show_mode(*MODES[2])  # Configurations (20 May)
```

```python
show_mode(*MODES[3])  # CLEA SAMD sweep (20 May)
```

```python
show_mode(*MODES[4])  # Ambient vs Only cabinets
```
